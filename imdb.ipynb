```python
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set_style("whitegrid")
sns.color_palette("Paired")
```

```python
sns.heatmap(data.corr(),annot=True)
plt.title('understanding the correlation')
```

```python
sns.boxplot(data=data,x="IMDB rating",y="Year of Release")
plt.title('boxplot for rating vs year of release')
```

```python
plt.figure(figsize=(12,12))
plt.title('Number of movies released per year recorded')
sns.countplot(y=data["Year of Release"])
plt.xticks(rotation=60)
plt.ylabel("year")
```

```python
sns.histplot(x=data["IMDB rating"],kde=True)
plt.title('histogram representing maximum IMDb rating count')
plt.xticks(rotation=90)
```

```python
plt.figure(figsize=(5,5))
plt.xticks(rotation=90)
plt.title("a scatter plot")
sns.scatterplot(x="Year of Release",y="IMDB rating",data=data)
```

```python
rating_cnt=data["IMDB rating"].value_counts()
```

```python
plt.figure(figsize=(12,6))
plt.title("Ratings ratio")
plt.pie(rating_cnt,labels=rating_cnt.index, autopct='%1.1f%%', startangle=180,labeldistance=1.15) 
plt.legend(loc="center right",bbox_to_anchor=(-1,1))
plt.show()
```
